```python
import torch
import numpy as np
import os
from sklearn.cluster import SpectralClustering
```

```python
sim_mat =  'monkeypox-adj'#'japan-adj'#'region-adj'#'state-adj-49'
path = os.path.join('..','data','{}.txt'.format(sim_mat))
adj  = torch.Tensor(np.loadtxt(open(path), delimiter=','))
print(adj)
```

```python
cluster_num = 3
sc = SpectralClustering(cluster_num, affinity='precomputed', n_init=100,
                         assign_labels='discretize')
cluster_rel = sc.fit_predict(adj)
print(cluster_rel)
institueion_list =[]
for i in range(cluster_num):
    institueion_list.append(np.where(cluster_rel==i)[0].tolist())

print(institueion_list)
```
